## 6.1 How data are generated

Up to this point, we’ve practiced understanding and manipulating data, and summarizing it with plots/statistics. But it's also important to understand where it *comes from* in the first place. And we're not just talking about the people or subjects the data is from. *Why* is someone's height 5'8", their eyes brown, their personality introverted? And *why* is there variation on these variables from person to person? Understanding this will be our first step into understanding the how we derive *answers* from data, instead of just pushing numbers around.

Let's go back to the example of height. Let's say we measure a person's height as 5'8", with no measurement error. Great! But what went into determining that this person's height would be this number specifically? We could probably think of things like their genes, how nutritious their diet is, or whether or not they've had any serious diseases. Let's try putting all that into a basic math equation:

```height = genes + nutrition + disease```

Maybe we study this process a bit more, and learn that some of these contributions have a bigger effect on height than others. Maybe, genes are twice as important as disease, which is twice as important as nutrition. So we can modify our equation to reflect each contributor's importance:

```height = 4*genes + 1*nutrition + 2*disease```

What we've just built is called a **model** how height data is generated. It is our best estimation of what the **data generation process** is. If it is accurate, then we can always know how tall someone will be by plugging gene, nutrition, and disease values into this equation. 

Sometimes we know the data generation process really well - we don't have to personally measure the gravity on Jupiter to know what it will be, because we already know the mathematical model that produces gravitational force:

$$gravitational force = G\frac{m_1*m_2}{r^2}$$

where *G* is the gravitational constant, m<sub>1</sub> is the mass of one body, m<sub>2</sub> is the mass of a second body, and r<sup>2</sup> is the distance between the centers of mass. We don't need to collect data about gravity on new planets so long as we know what their mass is.

However, most things in life are much more complicated than this. There are more variables that go into the calculation, and the relationships between variables are more complex. This is why some people nickname psychology as "the hardest science." We are very far away from perfectly modeling the exact data generation process for psychological constructs. 

## 6.2 Probability vs. statistics

Sometimes you can't perfectly compute the data generation process because of some physical limitations. E.g., when someone flips a coin, it will almost certainly land on either one side of the either. Which side it will be depends on how many flips the coin does and how high it flew, both of which are related to how forcefully it was flipped. But it's pretty impractical to measure the force of a finger flick. Instead, we can say 1) the result will be either heads OR tails, and 2) the coin is equally weighted, so both options are equally likely. We don't for sure know the outcome of any one particular flip (the result of each individual flip is **random**), but we know that over many flips about 50% will be heads and 50% will be tails. We can still write a model for this:

$$P(heads) = 0.5$$

which you can read as “the probability of heads is 0.5”. In the same way that percentages are numbers that range from 0% to 100%, probabilities are just numbers that range from 0 to 1. When using this **probability model** to answer the first question, we don’t actually know exactly what’s going to happen on a particular flip. Maybe we’ll get 10 heads in a row. But maybe we’ll get three heads. That’s the key thing: in probability theory, *the data generation model is known, but the data are not.*

So that’s probability. What about statistics? Statistical questions work the other way around. In statistics, we do not know the truth about the world - we don't have a model. All we have is the data, and it is from the data that we want to guess the truth about the world. Statistical questions tend to look more like these:

- If my friend flips a coin 10 times and gets 10 heads, are they playing a trick on me?
- If five cards off the top of the deck are all hearts, how likely is it that the deck was shuffled?
- If the lottery commissioner’s spouse wins the lottery, how likely is it that the lottery was rigged?

This time around, the only thing we have are data. What we know is that the friend flipped the coin 10 times and it came up heads every time. And what we want to **infer** is whether or not we should conclude that what we just saw was actually a fair coin being flipped 10 times in a row, or whether we should suspect that the friend is playing a trick. The data we have look like this:

$$H H H H H H H H H H H$$

and what we're trying to do is work out which “model of the world” probably generated that set of data. If the coin is fair, then the model we should adopt is one that says that the probability of heads is 0.5; that is,

$$P(heads) = 0.5$$

If the coin is not fair, then we should conclude that the probability of heads is not 0.5. I.e., 

$$P(heads) > 0.5$$

In other words, the statistical inference problem is to figure out which of these probability models is right. Clearly, the statistical question isn’t the same as the probability question, but they’re deeply connected to one another. Because of this, a good introduction to statistical theory will start with a discussion of what probability is and how it works.

## 6.3 The basic rules of probability

Before we get much further, we (perhaps unfortunately) do have to pause and define some mathematical rules about what a probability is. The meaning behind the math symbols is not particularly complex - these rules, if you put them into words, should seem pretty self-evident. They help us constrain what kinds of situations qualify as a probability.

First up, some definitions:

- The **sample space** is the set of possible outcomes for an experiment. We represent these by listing them within a set of squiggly brackets. For a coin flip, the sample space is ```{heads, tails}```. For a six-sided die, the sample space is each of the possible numbers that can appear: ```{1,2,3,4,5,6}```. For the amount of time it takes to get to work, the sample space is all possible real numbers greater than zero (since it can’t take a negative amount of time to get somewhere, at least not yet). We won’t bother trying to write out all of those numbers within the brackets.
- An **elementary event** is one outcome from the sample space. For example, this could be obtaining heads in a single coin flip, rolling a 4 on a throw of the die, or taking 21 minutes to get home by the new route.
- A **complex event** refers to multiple events in combination. E.g., the probability of two specific outcomes when two dice are rolled. In other words, this is the a set of multiple elementary events. 

Now that we have those definitions, we can outline the formal mathematical features of a probability, also known as **axioms**. These are the features that a value has to have if it is going to be a probability, under our definition of one. Let’s say that we have a sample space defined by N possible events, {E<sub>1</sub>, E<sub>2</sub>, E<sub>3</sub>, ... E<sub>N</sub>}, and X is a random variable denoting which of the events has occurred, where P(X = E<sub>i</sub>) is the probability of X equaling the i'th kind of possible event. Under this idea:

### Axioms 
1) **Probability cannot be negative**: P(X = E<sub>i</sub>) => 0

2) **The total probability of all outcomes in the sample space is 1**; that is, if we take the probability of each E<sub>i</sub> and add them up, they must sum to 1. We can express this using the summation symbol:

$$\sum_{i=1}^{N} P(X = E_i) = P(X = E_1) + P(X = E_2) + P(X = E_3) + ... + P(X = E_N) = 1$$

This is interpreted as saying “Take all of the N elementary events, which we have labeled from 1 to N, and add up their probabilities. These must sum to one.” Another way of saying this, which will come in handy later in the course, is that the probability of *NOT* E<sub>i</sub> is 1-P(X = E_i).

3) **The probability of any individual event cannot be greater than one**: P(X = E_i) <= 1. This is implied by the previous point; since they must sum to one, and they can’t be negative, then any particular probability cannot exceed one. 

Based on these axioms and some algebra, we can build additional mathematical rules that hold so long as the axioms are true:

### Mathematical rules
1) **Subtraction**: If the sum of all probabilities for a sample space is always 1 (axiom #2 above), then the probability of any particular event *NOT* occurring is equal to the sum of the probabilities of all other possible events (in R, the ```!``` symbol means "not," so we'll use it here too): 

$$P(!E_1) = 1 - P(E_1)$$

2) **Cumulative probability**: Say we roll a six-sided die, but we don't really care about the exact number that comes up. Instead, we care about rolling any of a subset of numbers - e.g., at least a 4 (so 4, or 5, or 6). In this case, we're writing a logical statement (>=4), and we're looking for the probability of that logical statement being true. For this, we add up the probabilities of each specific outcome that would be true under our logical statement:

$$P(E_i>=4) = P(E_4) + P(E_5) + P(E_6)$$

3) **AND events**: An AND event (also known as an "intersection" or referred to as "joint probability") is a type of complex event. This is the probability of both outcome 1 *AND* outcome 2 happening jointly across multiple events. (In R, ```&&``` is used to denote "and"): 

$$P(E_1 \&\& E_2) = P(E_1) * P(E_2)$$

4) **OR events**: An OR event (also known as a "union") is another complex event. This one is the probability of *either* outcome 1 *OR* outcome 2 happening across multiple events. (In R, we use two of the vertical slash symbols, ```||```): 

$$P(E_1 || E_2) = P(E_1) + P(E_2) - P(E_1 \&\& E_2)$$

In this rule, we need to remember to subtract the probability of A and B occuring together, because otherwise we're counting the outcome of A && B twice, since it's covered both by P(A) and P(B). Also, note that this is NOT the same thing as a cumulative outcome like rolling *at least a 4* on *one* die. This applies to multiple events happening. 


## 6.4 Probability distributions

As discussed in the last chapter, distributions show us the variability of a dataset. They are useful to us in probability theory too, because they can show us the range of event outcomes given an underlying probability model. For instance, let's say we have a fair six-sided die, which means the probability of rolling any of the numbers is 1/6 = 0.167. We can plot this to show, for every possible outcome of the die (on the x-axis), what the probability of that outcome is (on the y-axis). 

<img src="ch6-uniform.png"  width="400">

This looks a lot like a histogram when plotted this way, doesn't it? In fact, the above picture was generated using ```gf_dhistogram()```, just with the y-axis label changed from "density" to "probability." This reveals an inherent truth about probabilities - if we know the data generation process (the probability model), the proportions of values in any data generated by the process (the "density" in a density histogram) will match the expected probabilities (allowing for some fluctuation due to randomness). This is actually where special kinds of distributions get their names - a specific probability model always creates a specific distribution shape. 

Thus, even if we don't have a set of data to start with, when we know what the data generation process is, we can generate or **simulate** new data. We'll do this now to show you how certain types of probabilities correspond to certain probability distributions.

### Uniform distribution
First, let's tackle the example above of a fair six-sided die. There are 6 possible outcomes, ```{1,2,3,4,5,6}```. We know the probability corresponding to each of those values: ```{0.167,0.167,0.167,0.167,0.167,0.167}```. If we want to see a distribution for rolling this die 1,000,000 times, we can use the function ```sample()```:

````r
library(ggformula)

#possible values of the die
sample_space <- c(1,2,3,4,5,6)
#probabilities of each value
probs <- c(0.167,0.167,0.167,0.167,0.167,0.167)
#sample() takes a sample space vector, sample size, replace flag, and probability vector
outcomes <- sample(x=sample_space, size=1000000, replace=TRUE, prob = probs)
#put the outcomes into a dataframe for plotting
outcome_df <- data.frame(outcomes=outcomes)
#plot all the die rolls
gf_dhistogram(~ outcomes, data=outcome_df, color="gray", fill="blue", bins=6)
````

The ```sample()``` function in R is a general sampling function that lets you define any sample space and any corresponding probabilities. It takes four arguments: 1) a vector 'x' of possible values (i.e. the sample space); 2) how many values from that sample space you want to generate; 3) a flag to signal if you want to sample with replacement or not (i.e., once you roll a 6, is it still possible to roll another 6? If yes, replace=TRUE); and 4) a vector of the probabilities of each possible outcome (must be the same length as the sample space vector x). 

Examine the graph you made above. It should look like a rectangle, with roughly equal numbers of rolls in each possible die value. This is a good function to use if you know all the possible values when picking a sample, and know what probabilities they have. 

Next try the code below. Notice that the last argument, ```prob``` has been left out. What happens?

````r
#sample() function, but no prob argument
outcomes <- sample(x=sample_space, size=1000000, replace=TRUE)
#put the outcomes into a dataframe for plotting
outcome_df <- data.frame(outcomes=outcomes)
#plot all the die rolls
gf_dhistogram(~ outcomes, data=outcome_df, color="gray", fill="blue", bins=6)
````

Does it look pretty much the same? Why didn't the function show an error? 

This is because some functions have what's called "default arguments," where if you don't specify something (like for ```prob```), the underlying source code will default to a particular value. In this case, ```sample()``` has a default value of ```prob``` being all values have the same probability. When functions get complex like this, it's good to write down how it works in your R cheatsheet, or google for the documentation of the function. [Here](https://www.rdocumentation.org/packages/base/versions/3.6.2/topics/sample) is the documentation for ```sample()```.

Because a distribution like the uniform distribution, where every value is equally likely, is so common in the real world, R also includes a couple functions specifically for this distribution: ```runif()``` and ```dunif().``` The "unif" part of the function name is obvious - it refers to the uniform distribtion. In the first one, the "r" refers to generating a random sample from the uniform distribution, and in the second one, the "d" refers to calculating the density (or probability) of a particular value. 

````r
#sampling from the uniform distribution 
outcomes <- runif(n=1000000, min=0, max=6)
#putting sample into a dataframe for plotting
outcome_df <- data.frame(outcomes=outcomes)
#plot all the sampled values
head(outcome_df)
````

Note that the functions in R for these special distributions will sample continuous numbers, meaning any decimal value between the min and max arguments. So if we wanted to generate ordinal values like on a die, we need to round every generated number up to the nearest ordinal value. R gives a few functions for this: ```round()``` (round values to a specified decimal place), ```ceiling()``` (round up to the nearest integer), and ```floor()``` (round down to the nearest integer). We'll use ```ceiling()``` in our case to round up so all our values are only in {1,2,3,4,5,6}.  

````r
outcome_df$outcomes <- ceiling(outcome_df$outcomes)
head(outcome_df)
gf_dhistogram(~ outcomes, data=outcome_df, color="gray", fill="blue", bins=6)
````

Now try using ```dunif()``` to calculate the probability of rolling a 4, given a uniform probability distribution:

````r
# first argument is the value you want the probabilty of; second and third arguments are the min and max of the
# (continuous) distribution
dunif(4, min=0, max=6)
````

### Normal distribution

When a set of elementary events have equal probabilities of occuring, their probability distribution matches the special uniform distribution. This is the easiest distribution to understand in terms of the data generation process that creates it. However, there are a great many more distribution shapes that emerge because of more complicated underlying probability models. So in this course we're only going to introduce you to one more special distribution, the **normal distribution**. 

Imagine that it takes you, generally, 8 minutes to bike to class from your dorm room. However, some days you're really fast, so it takes only 7 or even 6. Other days, your tire pressure is low or you're feeling sluggish - it takes up to 10 minutes. Your mean travel time is 8, but there is some random fluctuation around that from day to day. 

In this case, you can calculate the probability of arriving at class in 8 minutes, but you don't do it the same way as calculating the probability of rolling a 1 on a die. This is because there are technically infinite possibilities of arrival time (maybe you get to class half an hour late!), and the extreme values are less likely to happen than the mean time of 8 minutes (lots of things would need to go wrong to make you half an hour late).    

As we will see many times in this course, a lot of data in the real world has this sort of underlying pattern, and is represented by the normal distribution. In order to generate a normal distribution, you can't specify it with a min and a max, since technically all values are possible. Instead, we describe a normal distribution based on its mean (the most probable outcome) and it's standard deviation (the distribution's spread). As you will recall from last chapter, 1 standard deviation means about 68% of the data in a normal distribution is 1 standard deviation above or below the mean, 95% is within 2 standard deviations, etc. In R, we'll use ```rnorm()``` and ```dnorm()``` for doing the same sorts of computations we did previously with the uniform distribution. For instance, imagine that your class biking time has a mean of 8 minutes and a standard deviation of 1:

````r
#sampling from the normal distribution 
outcomes <- rnorm(n=1000000, mean=8, sd=1)
#putting sample into a dataframe for plotting
outcome_df <- data.frame(outcomes=outcomes)
#plot all the sampled values
gf_dhistogram(~ outcomes, data=outcome_df, color="gray", fill="blue")
````

## 6.5 Conditional probability

Throughout this chapter, all the probability events we've discussed - rolling a die, time to get to class, etc. - had one thing in common: we only talked about one type of outcome at at time. We didn't consider any other variable that might have an effect on the event outcome. But there are lots of things in life where the outcome of one event affects the probability of another event ocurring. For instance, someone who gets sunburned is more likely to get skin cancer. We could calculate the raw probability of someone getting skin cancer, with no other information, but that might not be very useful for helping a particular individual assess their own risk. If we instead know how certain events impact others, we can factor those in to calculate more specific probabilities for other events. This is called **conditional probability**. 

Let’s take the 2016 US Presidential election as an example. There are two simple probabilities that we could use to describe the electorate. First, we know the probability that a voter in the US is affiliated with the Republican party: P(Republican) = 0.44 (since 44% of the electorate is registered as such). We also know the probability that a voter cast their vote in favor of Donald Trump: P(Trump) = 0.46. However, let’s say that we want to know the following: What is the probability that a person cast their vote for Donald Trump, *given that they are a Republican?*

To compute the conditional probability of A *given* B (written as P(A|B), with one vertical bar), we need to know the joint probability (that is, the probability of both A and B occurring) as well as the overall probability of B:

$$P(A|B) = \frac{P(A\&\&B)}{P(B)}$$

That is, we want to know the probability that both things are true, given that the one being conditioned upon is true.

It can be useful to think of this graphically. The figure below shows a flow chart depicting how the full population of voters breaks down into Republicans and Democrats, and how the conditional probability (conditioning on party) further breaks down the members of each party according to their vote.

<img src="ch6-condprob.png"  width="500">

As the equation above says, in order to compute a conditional probability, we need to know the joint probability of both events as well as the simple probability of the event being conditioned on. In another example, let’s say that we are interested in the following question: What is the probability that someone has diabetes, given that they are not physically active? – that is, P(diabetes|inactive). The table below summarizes a dataset with these two boolean variables, ```PhysActive``` and ```Diabetes```. The third column lists how many people in the dataset had each value on those variables. By dividing that number by the total n, we know the joint probability (4th column). 

| Diabetes   | PhysActive | n     | Joint probability |
| :--------: | :--------: | :---: | :---------------: |
| No         | No         | 2123  | 0.390             |
| No         | Yes        | 2770  | 0.509             |
| Yes        | No         | 349   | 0.064             |
| Yes        | Yes        | 201   | 0.037             |

For our conditional probability equation, the joint probability is the line in the table for probability that someone is both physically inactive ("No" on ```PhysActive```) and has diabetes ("Yes" on ```Diabetes```). The simple probability of ```PhysActive``` is the sum of probabilities that someone is inactive, regardless of their diabetes status. Thus:

$$P(diabetes|inactive) = \frac{0.064}{0.454} = 0.141$$

In other words, this is the proportion of people who are diabetic in just the subset of the data that are physically inactive. If this conditional probability of diabetes (taking inactivity into account) is different than the simple probability of diabetes, then we can conclude that diabetes and physically activity are likely related to each other - one has at least some sort of causal effect on the other, or another unmeasured variable has a causal effect on both. If instead, this conditional probability P(diabetes|inactive) is the same as the simple probability P(diabetes), then taking inactivity into account doesn't add any predictive value - the two variables are **independent**. 

$$Independence: P(A|B) = P(A)$$

Let's check that in our data - does P(diabetes|inactive) = P(diabetes)? The conditional probability as we already calculated is 0.141, while the simple probability of diabetes is (349+201)/5543 = 0.099. Those *seem* like different values... but maybe this is just due to some random variation in our data? Hang tight on that question - later in the course we'll learn some statistical tools to suggest how confident we can be that these probabilities are actually different!
